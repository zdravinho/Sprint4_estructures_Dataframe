```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('preu lloguer per trimestre i barri.csv')
freq = df['Preu'].value_counts()
freq = freq.sort_values(ascending=True)

# Creeu un gràfic de barres per visualitzar la freqüència de cada valor de la columna 'Districte'
sns.kdeplot(freq)

# Afegiu títol i etiquetes d'eixos

plt.title('Distribució del preu del lloguer')
plt.xlabel('Preu')
plt.ylabel('Densitat')
# Mostra el gràfic
plt.show()

plt.savefig('Variable_numerica.png')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('preu lloguer per trimestre i barri.csv')

# Convert Preu to numeric
df['Preu'] = pd.to_numeric(df['Preu'], errors='coerce')

# Drop NaN values
df = df.dropna()

# Create boxplot
sns.set(style='whitegrid')
fig, ax = plt.subplots(figsize=(8, 6))
sns.boxplot(data=df, x='Preu', y='Nom_Districte', palette='rocket', ax=ax)
sns.stripplot(data=df, x='Preu', y='Nom_Districte', jitter=0.2, color='black', size=3, ax=ax)
ax.set_xlabel('Preu (€)')
ax.set_ylabel('Districte')
plt.title('preu per districte')

# Addressing pitfalls
plt.tight_layout()  # Avoid overlapping labels and titles
plt.subplots_adjust(left=0.2)  # Make room for y-axis labels
plt.show()

plt.savefig('Variables_numerica_categorica.png')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('preu lloguer per trimestre i barri.csv')

# Convert Preu to numeric
df['Preu'] = pd.to_numeric(df['Preu'], errors='coerce')

# Drop NaN values
df = df.dropna()

# Calculate mean Preu for each Trimestre group
df_mean = df.groupby('Trimestre')['Preu'].mean().reset_index()

# Create line plot
sns.set(style='whitegrid')
fig, ax = plt.subplots(figsize=(8, 6))
sns.lineplot(x='Trimestre', y='Preu', data=df, ax=ax, color='black', label='Price')
sns.lineplot(x='Trimestre', y='Preu', data=df_mean, ax=ax, color='red', label='Mean Price')
ax.set_xlabel('Trimestre')
ax.set_ylabel('Preu (€)')
plt.title('Preu de casa per trimestre')
plt.legend()
plt.show()

plt.savefig('Trimestre_preu.png')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('preu lloguer per trimestre i barri.csv')

# Convert Preu to numeric
df['Preu'] = pd.to_numeric(df['Preu'], errors='coerce')

# Drop NaN values
df = df.dropna()

# Pivot the data to create a matrix
heatmap_data = df.pivot_table(index='Nom_Districte', columns='Trimestre', values='Preu')

# Create heatmap
sns.set(style='whitegrid')
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(heatmap_data, cmap='rocket', annot=True, fmt=".0f", linewidths=1, ax=ax)
ax.set_xlabel('Trimestre')
ax.set_ylabel('District')
plt.title('Housing Prices by Trimester and District')
plt.show()

plt.savefig('heatmap.png')
```
